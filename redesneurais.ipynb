```python
import pickle as pkl
with open('census.pkl', 'rb') as f:
    x_census_train, x_census_test, y_census_train, y_census_test = pkl.load(f)

with open('credit_risk_dataset.pkl', 'rb') as f:
    x_credit_train, x_credit_test, y_credit_train, y_credit_test = pkl.load(f)
```

```python
from sklearn.neural_network import MLPClassifier

creditNN = MLPClassifier(max_iter=1500, verbose =True, tol=0.00001, hidden_layer_sizes=(25,25), activation= 'logistic', solver='adam')

creditNN.fit(x_credit_train, y_credit_train)
```

```python
predict = creditNN.predict(x_credit_test)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_credit_test, predict))

```

```python

censusNN = MLPClassifier(max_iter=1000, verbose =True, tol=0.00001, hidden_layer_sizes=(55,55), activation= 'relu', solver='adam')

censusNN.fit(x_census_train, y_census_train)
    
```

```python
predict = censusNN.predict(x_census_test)
print(accuracy_score(y_census_test, predict))
```
